```python
import pandas as pd
from transformers import pipeline


```

```python
classifier = pipeline("text-classification", model="facebook/bart-large-mnli")
```

```python
claims = pd.read_csv("govt_claims.csv")
reports = pd.read_csv("ground_reality.csv")
```

```python
merged = pd.merge(claims, reports, on='scheme')
```

```python
def classify(row):
    text = f"{row['claim_text']} </s> {row['report_text']}"
    result = classifier(text, truncation=True)[0]
    return pd.Series([result['label'], result['score']])
```

```python
merged[['nli_label', 'confidence']] = merged.apply(classify, axis=1)
```
